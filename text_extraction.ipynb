```python
def get_data(url):
    headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
    }
    r = requests.get(url,headers=headers)
    return r.text
```

```python
def get_text(url):
    html_data = get_data(url)
    #print(html_data)
    soup = BeautifulSoup(html_data,'html.parser')
    data1 = []
    
    for data in soup.find_all("p"):
            data1.append(data.get_text())
        
    return data1
```

2 Analysis of Readability<br>
Analysis of Readability is calculated using the Gunning Fox index formula described below.<br>
Average Sentence Length = the number of words / the number of sentences<br>
Percentage of Complex words = the number of complex words / the number of words<br> 
Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)<br>

```python
import numpy as np
```

```python
def syllables_count(i):
    txt = processed_text(i)
    syllable_count=0
    for j in range(len(txt)):
        word=txt[j]
        for w in word:
            if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U') and word[-2:]!="es" and word[-2:]!="ed":
                    syllable_count=syllable_count+1
    return syllable_count
```

Complex words are words in the text that contain more than two syllables.

```python
def complex_word(i):
    txt = processed_text(i)
    count=0
    for j in txt:
        if syllables_count(j)>2:
            count+=1
    return count
```

Percentage of Complex words = the number of complex words / the number of words

7 Personal Pronouns:- 
To calculate Personal Pronouns mentioned in the text, we use regex to find the counts of the words - “I,” “we,” “my,” “ours,” and “us”. Special care is taken so that the country name US is not included in the list.

```python
import re
def personal_pronoun(i):
    pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    txt = processed_text(i)
    pronoun = []
    for l in range(len(txt)):
        pronouns.append(pronounRegex.findall(txt[l]))
    print(pronouns)
```
